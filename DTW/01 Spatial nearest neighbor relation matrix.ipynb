## 双向三维

```python
num = df1.shape[0]

dis300 = pd.DataFrame()
for i in range (num):
    dftemp = pd.DataFrame()
    x1 = df1.iloc[i,1]
    y1 = df1.iloc[i,2]
    z1 = df1.iloc[i,15]
    for j in range (num):
        x2 = df1.iloc[j,1]
        y2 = df1.iloc[j,2]
        z2 = df1.iloc[j,15]
        dis = ((x1-x2)**2+(y1-y2)**2+(z1-z2)**2)**(0.5)
        if dis <= 350 :print(i,",",j)
```
